```python
import pandas as pd
import numpy as np
```

```python
june = pd.read_csv('aggyellow_tripdata_2016-06.csv', parse_dates=True, index_col=0)
```

```python
may_data = pd.read_csv('aggyellow_tripdata_2016-05.csv', parse_dates=True, index_col=0)
```

```python
may = may_data.truncate(after=pd.Timestamp('2016-05-31 23:00:00')).mean(axis=0)
indexes = may[may < 5].index
may_data.drop(columns=indexes, inplace=True)
```

```python
with open('regions.txt', 'w') as f:
    for reg in may_data.columns.values:
        f.write(reg + '\n')
```

```python
june.drop(columns=indexes, inplace=True)
```

```python
june.to_csv('june_clear.csv')
```
